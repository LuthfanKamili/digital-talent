```python
def faktorial(n):
    i = 1
    result = 1
    while i<n:
        result=result*n
        n=n-1
    return result
```

```python
hasil = faktorial(5)
print (hasil)

hasil = faktorial (2) * faktorial(3)

print(hasil)
```

```python
import statistics

a = [0, 1, 1, 3, 4, 9, 15]

b = statistics.mean(a)
c = statistics.median(a)
d = statistics.mode(a)
e = statistics.stdev(a)
f = statistics.variance(a)

print(b,c,d,e,f)
```
