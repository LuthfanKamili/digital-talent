```python
import pandas as pd

gini_ratio=pd.read_csv('gini_ratio_2007-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
```

# 1. DESKRIPSI DATA


Gini Rasio atau koefisien rasio adalah alat mengukur derajat ketidakmerataan distribusi penduduk. Ini didasarkan pada kurva Lorenz, yaitu sebuah kurva pengeluaran kumulatif yang membandingkan distribusi dari suatu variable tertentu (misalnya pendapatan) dengan distribusi uniform (seragam) yang mewakili persentase kumulatif penduduk yang angkanya berkisar antara nol (pemerataan sempurna) hingga satu (ketimpangan yang sempurna).

Data ini berisi Rasio Gini per provinsi di Indonesia pada tahun 2011-2013. Penjelasan lebih detil mengenai Rasio Gini dapat diperoleh di tautan berikut: http://sirusa.bps.go.id/index.php?r=istilah/view&id=1461

Penjelasan mengenai Variabel pada Dataset ini:

kode_provinsi: Kode Provinsi

nama_provinsi: Nama Provinsi

tahun: Tahun

rasio_gini: Rasio Gini/Rasio Ketimpangan (jangkauan: 0 - 1)

# 2. Statistik Deskriptif

```python
min = gini_ratio.rasio_gini.min()
max = gini_ratio.rasio_gini.max()
med = gini_ratio.rasio_gini.median()
mea = gini_ratio.rasio_gini.mean()

print("Gini Rasio Terkecil :", min)
print("Gini Rasio Terbesar :", max)
print("Gini Rasio Rata-Rata :", mea)
```

```python
t = gini_ratio.groupby(['tahun']).mean()['rasio_gini']
print("Rata-Rata Gini Rasio per Tahun (2011-2013):")
print(t)
```

```python
s = gini_ratio.groupby(['nama_provinsi']).mean()['rasio_gini']
print ("Rata-Rata Gini Rasio per Provinsi (2011-2013):")
print (s)
```

# 3. Visualisasi Sederhana

```python
ax = gini_ratio.groupby('tahun').mean()['rasio_gini']
ax = ax.plot(kind='line', title ="Kenaikan Rasio Gini", figsize=(12,12), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Rasio_Gini", fontsize=12)
ax
```

```python
ax = gini_ratio.groupby('nama_provinsi').mean()['rasio_gini']
ax = ax.plot(kind='bar', title ="Rata-Rata Rasio Gini berdasarkan Provinsi selama 3 Tahun", figsize=(12,12), legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Rasio_Gini", fontsize=12)
ax
```

# 4. Narasi Analisa

### Berdasarkan data diatas dapat disimpulkan bahwa ketimpangan orang kaya dan miskin di tanah air masih tinggi. Hal ini tercemin dari data tahun 2010-2013, dimana rata-rata gini rasio di Indonesia selalu mengalami kenaikan dengan rata-rata 0.37. Dalam keaadaan ini, pemerintah harus segera melakukan tindakan antisipasi untuk mengurangi gap ratio yang semakin melebar. Salah satunya dengan menambahkan sejumlah kebijakan lain bagi pemerintah yakni meningkatkan layanan umum pada tingkat lokal terutama untuk provinsi dengan gini ratio diatas rata-rata.

### Jumlah provinsi dengan gini ratio diatas rata-rata adalah sebanyak 15 provinsi, yaitu:

```python
b = gini_ratio[['nama_provinsi','rasio_gini']].groupby('nama_provinsi').mean()
c = b.loc[(b['rasio_gini']>mea)]
print(c)
```

###  Provinsi dengan rata-rata rasio tertinggi, yaitu

```python
d = gini_ratio[['nama_provinsi','rasio_gini']].groupby('nama_provinsi').max()
e = d.loc[(d['rasio_gini']==max)]
print(e)
```

###  Sedangkan untuk provinsi dengan gini ratio di bawah rata-rata, harus tetap melakukan perbaikan-perbaikan untuk menurunkan gini ratio, karena meskipun rendah namun trend gini ratio mengalami peningkatan tiap tahunnya.
### Provinsi dengan rata-rata rasio terendah, yaitu

```python
f = gini_ratio[['nama_provinsi','rasio_gini']].groupby('nama_provinsi').min()
g = f.loc[(f['rasio_gini']==min)]
print('4.Provinsi dengan rata-rata rasio terendah, yaitu')
print(g)
```
